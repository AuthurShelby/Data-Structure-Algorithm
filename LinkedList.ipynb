```python
class Node:
    
    def __init__(self,data):
        self.data = data 
        self.next = None

class LinkedList:

    def __init__(self):
        self.head = None
    
    # for displaying the node

    def display(self):
        temp = self.head
        while temp!=None:
            print(temp.data , end=' -> ')
            temp = temp.next
        print('None')
    
    # for appending a new element at the end  

    def append(self):
        temp = self.head
        while temp.next!=None:
            temp = temp.next
        temp.next = Node(input('Enter a new node: ').strip())
    
    # for counting the number of nodes in the linked list 

    def count(self):

        temp = self.head
        count = 0 
        while temp!=None:
            count+=1
            temp = temp.next 
        print(f'Number of Nodes {count}')
    
    # for searching an element in the linked list 
    def search(self):

        temp = self.head
        Position = 0
        target = input('Enter the target data: ').strip()
        Found = False # Flag
        while temp!=None:
            Position +=1
            if temp.data == target:
                print(f'Target found at node {Position}')
                Found = True # Flag Set to True  
                break
            temp = temp.next
        # If the flag is still False
        if not Found:
            print(f'Sorry target data {target} not found!')
        
    # sorting the linked list using selection sort 

    def sort(self):

        swapped = True
        while swapped:
            swapped = False
            current = self.head
            while current.next != None:
                if current.data > current.next.data: 
                    current.data , current.next.data = current.next.data , current.data
                    swapped = True
                current = current.next 
    
    # for inserting an element at any position

    def insert(self):
        
        temp = self.head
        target = input('Enter a data to search -> ').strip()
        while temp!=None:
            if target == temp.data:

                # New node to be created
                NewNode = Node(input('New Node data : ')) 
                # assigning the address of the next node to the new node 
                NewNode.next = temp.next
                # assigning the address of the new node to previous node
                temp.next = NewNode
                print('Inserted!')
                break

            temp = temp.next

        if temp == None:
            print('Target not found!')

    # for removing an element any position

    def delete(self):

        if self.head is None:
            print('Linked-List not found!')
            return 
            
        
        prev = None
        temp = self.head

        target = input('Enter the data to be removed: ').strip()
        
        # target is the first node 
        if target==temp.data: 
            self.head = self.head.next
            print('removed!')
            return 

        while temp!=None and target!=temp.data :
            prev = temp
            temp = temp.next
    
        # if target not found
        if temp is None:
            print(f'Target {target} not found!')
            return 
        
        # target data is removed 
        prev.next = temp.next
        print('removed!')

    # Reverse the linked list 

    def reverse(self):
        # Initializing the three pointers previous , current and next 
        prev = None
        curr = self.head
        next = None
        while curr!=None:
            next = curr.next  # for getting the address of the next node else the address will be lost  
            curr.next = prev  # shifting of the pointers starts
            prev = curr 
            curr = next

        # assigning the head as prev  
        self.head = prev
        
        
```

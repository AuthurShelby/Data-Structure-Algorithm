# Queue Implementation Using Linked List
In a linked list-based implementation, the queue is implemented using a linked list where each node points to the next node. This avoids the limitation of a fixed size as seen in array-based queues. In this case, we maintain:

- Front: Points to the first node.

- Rear: Points to the last node.

```python
'''
This is the Linked List implementation of Queue (Menu - Driven)
'''

class Node:
    def __init__(self ,data):
        self.data = data 
        self.next = None 
    
class Queue:
    
    def __init__(self):
        self.front = None 
        self.rear = None

    # To check if it is empty 
    def isEmpty(self):
        return self.front is None
    
    # To add elements at the rear
    def Enqueue(self , element):
        newElement = Node(element)
        if self.isEmpty():
            self.front = self.rear = newElement
        else:
            self.rear.next =  newElement 
            self.rear = newElement
        print(f'Value {element} Enqueued!')
    
    # To remove the first/Top element
    def Dequeue(self):
        if self.isEmpty():
            print('The queue is empty!')
        else:
            dqElement = self.front.data
            self.front = self.front.next 
            if self.front is None:
                self.rear = None
            print(f'Value {dqElement} Dequeued!')
        
    # To get the first element in the queue
    def peek(self):
        if self.isEmpty():
            print('The queue is empty!')
        else:
            print(f'The Front element is {self.front.data}.')
    
    # To Display Whole Queue
    def display(self):
        if self.isEmpty():
            print('The queue is empty!')
        else:
            temp = self.front 
            print('Current Queue:' , end=' ')
            while temp:
                print(temp.data , end='->')
                temp = temp.next
            print("None")
    
# Menu-driven interface
def main():
    queue = Queue()

    while True:
        print("\n----- Queue Menu -----")
        print("1. Enqueue")
        print("2. Dequeue")
        print("3. Peek")
        print("4. Display Queue")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == '1':
            element = input("Enter the element to enqueue: ")
            queue.Enqueue(element)
        elif choice == '2':
            queue.Dequeue()
        elif choice == '3':
            queue.peek()
        elif choice == '4':
            queue.display()
        elif choice == '5':
            print("Exiting...")
            break
        else:
            print("Invalid choice! Please enter a number between 1 and 5.")

if __name__ == "__main__":
    main()
```

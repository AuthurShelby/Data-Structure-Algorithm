# This is a linear queue, meaning:
- Once rear reaches size, even if you Dequeue() items, the queue is considered full.
- This leads to inefficient use of memory if you're doing multiple enqueues and dequeues.
- So The solution is to Make the queue circular or implement using Linked List.

```python
'''
This is the array implementation of Queue.
'''

class Queue:

    def __init__(self , size):
        
        self.size = size # size of the queue  
        self.queue = [None]*self.size # Initializing the queue with a fixed length
        self.front = 0 
        self.rear = 0  

    # To check if the queue is Empty 
    def isEmpty(self):
        return self.front == self.rear
    # To check if the queue is Full
    def isFull(self):
        return self.size == self.rear

    # To Insert an element
    def Enqueue(self , element):
        if self.isFull():
            print('The queue is Full!')
        else:
            self.queue[self.rear] = element
            self.rear += 1
            print(f'Value {element} Enqueued!')

    # To Remove the front element
    def Dequeue(self):
        if self.isEmpty():
            print('The queue is empty!')
        else:
            dqElement = self.queue[self.front]
            self.front += 1
            print(f'Value {dqElement} Dequeued!')
    
    # To get the Front/First element
    def peek(self):
        if self.isEmpty():
            print('The queue is empty!')
        else:
            FRONT = self.queue[self.front] 
            print(f'The front value is {FRONT}')

def main():
    # Get the size of the queue from the user
    size = int(input("Enter the size of the queue: "))
    queue = Queue(size)

    while True:
        print("\n--- Menu ---")
        print("1. Enqueue")
        print("2. Dequeue")
        print("3. Peek")
        print("4. Exit")

        choice = int(input("Enter your choice: "))

        if choice == 1:
            element = int(input("Enter the element to enqueue: "))
            queue.Enqueue(element)
        elif choice == 2:
            queue.Dequeue()
        elif choice == 3:
            queue.peek()
        elif choice == 4:
            print("Exiting the program.")
            break
        else:
            print("Invalid choice! Please choose again.")
if __name__ == '__main__':
    main()
```

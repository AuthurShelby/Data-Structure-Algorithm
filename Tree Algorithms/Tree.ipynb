```python

#             R (0)
#           /    \ 
#        A (1)     B (2)
#      /    \     /   \
#    C (3)  D(4)  E(5)  F(6)

# array implementation 

def left_node(arr , index):
    ''' For fetching all the left nodes in the graph '''
    if index >= len(arr):
        return 
    
    left_index = 2 * index + 1
    right_index = 2 * index + 2

    if left_index < len(arr):
        print('left node = ', arr[left_index])
    
    left_node(arr , left_index)
    left_node(arr , right_index)

def right_node(arr , index):
    ''' For fetching all the rigt nodes in the graph '''
    if index >= len(arr):
        return 

    right_index = 2 * index + 2
    left_index = 2 * index + 1

    if right_index < len(arr):
        print('right node = ' , arr[right_index])
    
    right_node(arr , right_index)
    right_node(arr , left_index)


def print_tree(arr , index=0):
    ''' For printing the whole graph '''

    if  index >= len(arr):
        return 
    
    # to get the current index
    node = arr[index]

    right_index = 2 * index + 2
    left_index = 2 * index + 1 

    # Get left and right children if they exist
    left_child = arr[left_index] if left_index < len(arr) else None
    right_child = arr[right_index] if right_index < len(arr) else None

    print(f'{node} -> Left child : {left_child} -- Right Child : {right_child}')

    # Recurse to left and right subtree
    print_tree(arr , left_index)
    print_tree(arr , right_index)

if __name__ == '__main__':

    binary_tree_array = ['R', 'A', 'B', 'C', 'D', 'E', 'F']
    print('-- Left Nodes --')
    left_node(binary_tree_array , 0)
    print('-- Right Node --')
    right_node(binary_tree_array , 0)
    print('-- Whole Tree --')
    print_tree(binary_tree_array , 0)
```

```python
#          1
    #    /   \
    #  2       3
    # / \     / \
#    4   5   6   7

# implementation using Linked List

from collections import deque


# Node class
class Node:
    def __init__(self,data):
        self.data = data
        self.left = None 
        self.right = None
    
class BinaryTree:

    def __init__(self):
        self.root = None
    

    # for inserting a new node in the tree
    def insert(self):
        # if the root is empty or tree is not present
        if self.root is None:
            rootNode = int(input('Enter a value for the root : '))
            self.root = Node(rootNode)
            return 
        
        queue = deque([self.root])

        while queue:
            current = queue.popleft()
            
            if not current.left :
                LeftNode = int(input('Enter the left node value: ')) 
                current.left = Node(LeftNode)
                print('Inserted the left node...') 
                return 
            else:
                queue.append(current.left)

            if not current.right:
                RightNode = int(input('Enter the right node  value: '))
                current.right = Node(RightNode)
                return 
            else:
                queue.append(current.right)
        
    # left ->  root -> right 
    def inorder(self , node):
        if node:
            # starts from the left
            self.inorder(node.left)
            print(node.data , end=' ')
            # ends at the right
            self.inorder(node.right)

    # root -> left -> right
    def preorder(self , node):
        if node: 
            print(node.data , end=' ')
            # first left
            self.preorder(node.left)
            # then right
            self.preorder(node.right)

    # left -> right -> root
    def postorder(self , node):
        if node:
            # checks all the left
            self.postorder(node.left)
            # then to the right
            self.postorder(node.right)
            print(node.data , end=' ')
  
    # level order
    def levelorder(self , node):

        if not node:
            return 

        queue = deque([node])

        while queue:
            nodes = queue.popleft()
            print(nodes.data , end=' ')

            if nodes.left : 
                queue.append(nodes.left)

            if nodes.right:
                queue.append(nodes.right)


# Menu-Driven
# MAIN
if __name__ == "__main__":
    BT = BinaryTree()

    while True:
        print("\n--- Binary Tree Menu ---")
        print("1. Insert a node")
        print("2. Inorder traversal")
        print("3. Preorder traversal")
        print("4. Postorder traversal")
        print("5. Level order traversal")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            BT.insert()
        elif choice == '2':
            print("Inorder Traversal: ", end='')
            BT.inorder(BT.root)
            print()
        elif choice == '3':
            print("Preorder Traversal: ", end='')
            BT.preorder(BT.root)
            print()
        elif choice == '4':
            print("Postorder Traversal: ", end='')
            BT.postorder(BT.root)
            print()
        elif choice == '5':
            print("Level Order Traversal: ", end='')
            BT.levelorder(BT.root)
            print()
        elif choice == '6':
            print("Exiting program. Bye!")
            break
        else:
            print("Invalid choice! Please try again.")
```
